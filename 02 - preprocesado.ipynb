# Encoding

```python
import pandas as pd

"""
Este código realiza el preprocesamiento de un dataset con datos categóricos y ordinales.
1. Se codifican las columnas ordinales, respetando el orden de las categorías definidas. Esto incluye:
   - 'ESTU_VALORMATRICULAUNIVERSIDAD', que refleja rangos de valores de matrícula.
   - 'FAMI_ESTRATOVIVIENDA', que refleja el estrato socioeconómico del hogar.
   - 'ESTU_HORASSEMANATRABAJA', que refleja el número de horas que el estudiante trabaja a la semana.

2. Se aplicará **one-hot encoding** a las columnas nominales, es decir, aquellas que no tienen un orden natural:
   - 'ESTU_PRGM_ACADEMICO', 'ESTU_PRGM_DEPARTAMENTO', 'FAMI_TIENEINTERNET',
     'FAMI_EDUCACIONPADRE', 'FAMI_EDUCACIONMADRE', 'ESTU_PAGOMATRICULAPROPIO', 'RENDIMIENTO_GLOBAL'.

3. Las columnas 'ID' y 'PERIODO' no se codifican ni transforman, ya que son necesarias para el análisis, pero se mantienen en el dataset.

4. Finalmente, los valores booleanos resultantes de `pd.get_dummies()` (utilizado para one-hot encoding) se convierten a enteros (0 y 1) para que puedan ser utilizados en un modelo de machine learning.
"""

# Codificación de columnas ordinales

# Definir el orden para las categorías de 'ESTU_VALORMATRICULAUNIVERSIDAD'
valores_matricula = ['Menos de 500 mil', 'Entre 500 mil y menos de 1 millón',
                     'Entre 1 millón y menos de 2.5 millones', 'Entre 2.5 millones y menos de 4 millones',
                     'Entre 4 millones y menos de 5.5 millones', 'Entre 5.5 millones y menos de 7 millones',
                     'Más de 7 millones', 'No pagó matrícula']
df['ESTU_VALORMATRICULAUNIVERSIDAD'] = pd.Categorical(df['ESTU_VALORMATRICULAUNIVERSIDAD'],
                                                      categories=valores_matricula, ordered=True)
df['ESTU_VALORMATRICULAUNIVERSIDAD'] = df['ESTU_VALORMATRICULAUNIVERSIDAD'].cat.codes

# Definir el orden para 'FAMI_ESTRATOVIVIENDA'
estrato_vivienda = ['Estrato 1', 'Estrato 2', 'Estrato 3', 'Estrato 4', 'Estrato 5', 'Estrato 6', 'Sin Estrato']
df['FAMI_ESTRATOVIVIENDA'] = pd.Categorical(df['FAMI_ESTRATOVIVIENDA'], categories=estrato_vivienda, ordered=True)
df['FAMI_ESTRATOVIVIENDA'] = df['FAMI_ESTRATOVIVIENDA'].cat.codes

# Definir el orden para 'ESTU_HORASSEMANATRABAJA'
horas_trabajo = ['0', 'Menos de 10 horas', 'Entre 11 y 20 horas', 'Entre 21 y 30 horas', 'Más de 30 horas', 'Desconocido']
df['ESTU_HORASSEMANATRABAJA'] = pd.Categorical(df['ESTU_HORASSEMANATRABAJA'], categories=horas_trabajo, ordered=True)
df['ESTU_HORASSEMANATRABAJA'] = df['ESTU_HORASSEMANATRABAJA'].cat.codes

# Lista de columnas nominales para aplicar one-hot encoding (excluyendo las ordinales ya codificadas)
nominal_columns = ['ESTU_PRGM_ACADEMICO', 'ESTU_PRGM_DEPARTAMENTO', 'FAMI_TIENEINTERNET',
                   'FAMI_EDUCACIONPADRE', 'FAMI_EDUCACIONMADRE', 'ESTU_PAGOMATRICULAPROPIO', 'RENDIMIENTO_GLOBAL']

df_encoded = pd.get_dummies(df, columns=nominal_columns)

# Convertir los valores True/False a 0/1
df_encoded = df_encoded.astype(int)

# Verificar los primeros resultados y dimensiones del dataset final
print(df_encoded.head())
print(f"El dataset tiene {df_encoded.shape[0]} filas y {df_encoded.shape[1]} columnas.")
```
